```python
data_id=pd.read_csv('/content/gdrive/My Drive/hackathon/hackathon-upload/Hackathon_Ideal_Data.csv',sep=',')
data_id_july=data_id.loc[data_id['PERIODCODE']=='7/1/2019']
data_id_july1=data_id_july[['GRP','SGRP','SSGRP','CMP','MBRD','BRD']]
data_val2=data_val1[['GRP','SGRP','SSGRP','CMP','MBRD','BRD']]
data_val3 = pd.merge(data_val2, data_id_july1, on=['GRP','SGRP','SSGRP','CMP','MBRD','BRD'], how='left', indicator='july')
data_val3['july'] = np.where(data_val3['july'] == 'both', np.nan, 0)
data_val3
```

```python
data_val_new.to_csv('/content/gdrive/My Drive/hackathon/hackathon-upload/code2-july/data_val_new.csv') 
data_val_new.isna().sum(axis = 0)
data_w_jul3= data_w_jul[['STORECODE','DAY','VALUE']].groupby(['STORECODE','DAY']).sum()
data_w_jul3.to_csv('/content/gdrive/My Drive/hackathon/hackathon-upload/code2-july/data_w_jul3.csv')
data_w_jul4=data_w_jul3.groupby(['STORECODE']).size().reset_index(name='Count')
data_w_jul5= data_w_jul[['TOWN','STORECODE','OUTLET_TYPE','GRP','SGRP','SSGRP','CMP','MBRD','BRD','SKU','VALUE']].groupby(['TOWN','STORECODE','OUTLET_TYPE','GRP','SGRP','SSGRP','CMP','MBRD','BRD','SKU']).sum()
```

```python
data_w_jul5.to_csv('/content/gdrive/My Drive/hackathon/hackathon-upload/code2-july/data_w_jul5.csv') 
data_w_jul5=pd.read_csv('/content/gdrive/My Drive/hackathon/hackathon-upload/code2-july/data_w_jul5.csv')
data_w_jul6=pd.merge(data_w_jul5,data_w_jul4, on=['STORECODE'])
data_w_jul6["TOTAL_VALUE_31"]=data_w_jul6['VALUE']/data_w_jul6['Count']*31
data_w_jul6.drop(['VALUE','Count'], axis = 1, inplace=True)
data_w_jul6.rename(columns = {'TOTAL_VALUE_31':'july'}, inplace = True)
data_w_jul6
```

```python
#data_val_new2= pd.read_csv('/content/gdrive/My Drive/hackathon/hackathon-upload/code2-july/data_val_new2.csv') 
GRP= list(data_val_new2.GRP.unique())
len(GRP)
```

```python
GRP1=GRP[:30]
GRP2=GRP[30:165]
```

```python
data_val_new3= data_val_new2[['GRP','SGRP','SSGRP','CMP','MBRD','BRD','july']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor

final1=pd.DataFrame(columns=['GRP','SGRP','SSGRP','CMP','MBRD','BRD','7/1/2019'])
for i in GRP1:
    data_val_new4=data_val_new3[data_val_new3['GRP']==i]
    dummy_variables = ['GRP','SGRP','SSGRP','CMP','MBRD','BRD']
    for var in dummy_variables:
        dummy = pd.get_dummies(data_val_new4[var], prefix = var, drop_first = False)
        train = pd.concat([data_val_new4, dummy], axis = 1)
    train = train.drop(dummy_variables, axis = 1)
    train_1 = train[pd.isnull(train['july']) == False]
    train_2 = train[pd.isnull(train['july'])]
    
    X_1=train_1.iloc[:,1:].values
    y_1=train_1.iloc[:,0].values
    X_2=train_2.iloc[:,1:].values
    y_2=train_2.iloc[:,0].values
    
    gbr = GradientBoostingRegressor(n_estimators=100,max_depth=10, random_state=0)
    
    gbr.fit(X_1, y_1)

    generatedValues = gbr.predict(X_2)
    train_2['july'] = generatedValues
    data = train_1.append(train_2)
    data.sort_index(inplace=True)
    data_val_new4["7/1/2019"]=data['july']
    data_val_new4.drop(['july'], axis = 1, inplace=True)
    final1=final1.append(data_val_new4)
final1
```

```python
final1.to_csv('/content/gdrive/My Drive/hackathon/hackathon-upload/code2-july/final1.csv',index=False)
```

```python
final2=pd.DataFrame(columns=['GRP','SGRP','SSGRP','CMP','MBRD','BRD','7/1/2019'])
for i in GRP2:
    data_val_new4=data_val_new3[data_val_new3['GRP']==i]
    dummy_variables = ['GRP','SGRP','SSGRP','CMP','MBRD','BRD']
    for var in dummy_variables:
        dummy = pd.get_dummies(data_val_new4[var], prefix = var, drop_first = False)
        train = pd.concat([data_val_new4, dummy], axis = 1)
    train = train.drop(dummy_variables, axis = 1)
    train_1 = train[pd.isnull(train['july']) == False]
    train_2 = train[pd.isnull(train['july'])]
    
    X_1=train_1.iloc[:,1:].values
    y_1=train_1.iloc[:,0].values
    X_2=train_2.iloc[:,1:].values
    y_2=train_2.iloc[:,0].values
    
    gbr = GradientBoostingRegressor(n_estimators=100,max_depth=10, random_state=0)
    
    gbr.fit(X_1, y_1)

    generatedValues = gbr.predict(X_2)
    train_2['july'] = generatedValues
    data = train_1.append(train_2)
    data.sort_index(inplace=True)
    data_val_new4["7/1/2019"]=data['july']
    data_val_new4.drop(['july'], axis = 1, inplace=True)
    final2=final2.append(data_val_new4)
final2
```
